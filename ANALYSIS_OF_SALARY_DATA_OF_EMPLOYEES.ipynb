```python
# Import Pandas library
import pandas as pd
```

```python
# Import dataset from Computer into Google Colab
from google.colab import files
uploaded = files.upload()
```

```python
# Load dataset into dataframe
df = pd.read_csv('Salary_Data.csv')
df.head()
```

```python
# Check for count of rows and columns
df.shape
```

```python
# Use this function to check data that will be relevant in our analysis for any data entry error
df ['Education Level'].value_counts()
```

```python
# Change 'Master's' to 'Master's Degree' in 'Education Level' column
df['Education Level'] = df['Education Level'].replace({"Master's": "Master's Degree"})
```

```python
df['Education Level'].value_counts()
```

```python
# Change 'Bachelor's' to 'Bachelor's Degree' in 'Education Level' Column
df['Education Level'] = df['Education Level'].replace({"Bachelor's":"Bachelor's Degree"})
```

```python
df['Education Level'].value_counts()
```

```python
# Change 'phD' to 'PhD' in 'Education Level' Column
df['Education Level'] = df['Education Level'].replace({"phD":"PhD"})
```

```python
df['Education Level'].value_counts()
```

```python
# Check for duplicates
df.duplicated().sum()
```

```python
# Delete all duplicates
df = df.drop_duplicates()
```

```python
# Check if all duplicates have been deleted
df.duplicated().sum()
```

```python
# Check for null values
df.isnull().sum()
```

```python
# Drop the null cells
df = df.dropna()
```

```python
# Check if null cells have been removed
df.isnull().sum()
```

```python
# Basic Descriptive Statistics
mean = df['Salary'].mean()
median = df['Salary'].median()
mode = df['Salary'].mode()
std = df['Salary'].std()

# Format the results
format_mean = "{:.2f}".format(mean)
format_median = "{:.2f}".format(median)
format_std = "{:.2f}".format(std)

print("Mean:", format_mean)
print("Median:", format_median)
print("Mode:", mode)
print("Standard Deviation:", std)
```

```python
# Get mean salary for males
male_mean = df[df['Gender'] == 'Male']['Salary'].mean()

# Get mean salary for females
female_mean = df[df['Gender'] == 'Female']['Salary'].mean()

#Display results
print("Mean salary for males:", male_mean)
print("Mean salary for females:", female_mean)
```

```python
# Store the cleaned dataset into a new csv file
df.to_csv('Salary_Data_Cleaned.csv', index=False)
```

```python
# Import matplotlib
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt

# Get the mean salary for each Gender
mean_salary = df.groupby('Gender')['Salary'].mean()

# Plot the bar chart
mean_salary.plot(kind='bar', color=['blue', 'pink', 'red'])

# Customize the chart
plt.title('Mean Salary by Gender')
plt.xlabel('Gender')
plt.ylabel('Mean Salary')
plt.xticks(rotation=0)

# Display the plot
plt.show()
```

```python
# Group by 'Job title' and calculate their mean salary
mean_salary_by_job = df.groupby('Job Title')['Salary'].mean()

# Sort by values in descending order and select top 5
top_5_jobs = mean_salary_by_job.sort_values(ascending=False).head(5)

# Plot the bar chart
top_5_jobs.plot(kind='bar', color='pink')

#Customize the chart
plt.title('Top 5 Average Salaries per Job Title')
plt.xlabel('Job Title')
plt.ylabel('Average Salary')
plt.xticks(rotation=45)

# Display the plot
plt.show()

```

```python
# Group by 'Job Title' and calculate their mean salary
mean_salary_by_job = df.groupby('Job Title')['Salary'].mean()

# Sort by values in descending order and select top 5
top_5_jobs = mean_salary_by_job.sort_values(ascending=False).head(5)

# Plot the horizontal bar chart
top_5_jobs.plot(kind='barh', color=['pink', 'green', 'orange', 'blue','red'])

# Customize the chart
plt.title('Top 5 Average Salaries per Job Title')
plt.xlabel('Average Salary')
plt.ylabel('Job Title')

# Display the plot
plt.show()
```

```python
# Create a scatter plot for Salary distribution across Ages

# Plot the scatter chart
plt.scatter(df['Age'], df['Salary'])

# Customize the chart
plt.title('Salary distribution across Ages')
plt.xlabel('Age')
plt.ylabel('Salary')

# Display the plot
plt.show()
```

```python
# Create a chart showing the Average Salary for each Education Level

# Group by 'Education level' and calculate their mean salary
df.groupby('Education Level')['Salary'].mean().plot(kind='bar', color=['red','blue','green','purple'])

# Customize the chart
plt.title('Average Salary by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Average Salary')

# Display the plot
plt.show()
```

```python

# Create a chart showing the Average Salary for each Education Level

# Group by 'Education level' and calculate their mean salary
df.groupby('Education Level')['Salary'].mean().plot(kind='line', color=['red','blue','green','purple'])

# Customize the line chart
plt.title('Average salary by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Average Salary')

# Display the plot
plt.show()
```
